```python
#Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Rodando o DF 'NetFlix'
NF_df = pd.read_csv("/content/netflix_titles.csv")
NF_df
```

```python
#Rodando o DF 'Rotten Tomatoes'
RT_df = pd.read_csv("/content/rotten_tomatoes_movies.csv")
RT_df
```

```python
#Rodando DF IMDB
IMDB_df = pd.read_csv("/content/imdb_top_1000.csv")
IMDB_df
```

##Questão 1
***
###Podemos concluir que o filme de gênero "Drama" melhor avaliado é: "The Breaking Point". Já o filme pior avaliado é: "A Beautiful Life" de acordo com fontes do Rotten Tomatoes

```python
#Filtramos o gênero dos filmes por "Drama"
drama_df = RT_df.loc[RT_df['genres'] == 'Drama']


#Descobrinco a nota máxima e mínima
nota_alta = drama_df['tomatometer_rating'].max()
nota_baixa = drama_df['tomatometer_rating'].min()


#Achamos o nome do filme de acordo com a nota máxima e mínima encontrada
melhor_drama = drama_df.loc[drama_df['tomatometer_rating'] == 100.0, 'movie_title'].values[0]
print(f"O melhor filme do gênero Drama é: {melhor_drama}, com uma nota de {nota_alta}  ")

print("------------------------------------------------")

pior_drama = drama_df.loc[drama_df['tomatometer_rating'] == 0.0, 'movie_title'].values[0]
print(f"O melhor filme do gênero Drama é: {pior_drama}, com uma nota de {nota_baixa}  ")
```

##Questão 2
***
###Feita a analise, concluimos que o filme de comédia melhor avaliado é: "Kenny". Já o filme pior avaliado é "Going The Distance" de acordo com fontes do Rotten Tomatoes

```python
#Filtramos o gênero dos filmes por "Comedia"
comedia_df = RT_df.loc[RT_df['genres'] == 'Comedy']

#Descobrinco a nota máxima e mínima
nota_alta_comedia = comedia_df['tomatometer_rating'].max()
nota_baixa_comedia = comedia_df['tomatometer_rating'].min()

#Achamos o nome do filme de acordo com a nota máxima e mínima encontrada
melhor_comedia = comedia_df.loc[comedia_df['tomatometer_rating'] == 100.0, 'movie_title'].values[0]
print(f"O melhor filme do gênero Comedia é: {melhor_comedia}, com uma nota de {nota_alta_comedia}  ")

print("------------------------------------------------")

pior_comedia = comedia_df.loc[comedia_df['tomatometer_rating'] == 0.0, 'movie_title'].values[0]
print(f"O pior filme do gênero Comedia é: {pior_comedia}, com uma nota de {nota_baixa_comedia}  ")
```

##Questão 6
***
###A quantidade de filmes com classificação TV-14 na netflix é de 2160. Porém é possivel notar através do gráfico que em sua maioria os filmes tem classificação TV-MA, ou seja, muitos filmes não são recomendados para menores de 17 anos, diferente da segunda maior classificação (TV-14) que não é recomendado para menores de 14 anos. Isso nos leva a crer que no catalogo da NetFlix a maioria dos filmes é +18

```python
qnt_classificacao = (NF_df['rating'] == 'TV-14').sum()
print(f"A quantidade de filmes na classificação TV-14 na NetFlix é de: {qnt_classificacao}")

# Calculando a quantidade de cada classificação
contagem_classificacao = NF_df['rating'].value_counts()

# Criando o gráfico de barras
plt.figure(figsize=(10, 6))
plt.bar(contagem_classificacao.index, contagem_classificacao.values)
plt.xlabel("Classificação")
plt.ylabel("Quantidade")
plt.title("Quantidade de Filmes por Classificação")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Convert 'original_release_date' to datetime objects
RT_df['original_release_date'] = pd.to_datetime(RT_df['original_release_date'])

# Extract the year
RT_df['release_year'] = RT_df['original_release_date'].dt.year

# Create the regression plot
sns.regplot(x='release_year', y='tomatometer_rating', data=RT_df.sample(1000), line_kws={"color": "red"})
plt.show()
```

##Questão 9
***
###De acordo com a analise realizada podemos entender como os gêneros de filmes de distribuem entre as diferentes classificações, podemos observar que as classificações de maior predominância na NetFlix (TV-MA e TV-14) assistem filmes internacionais.

```python
# Agrupa os filmes por classificação e obtém o gênero mais frequente em cada grupo
def most_frequent_genre(group):
  genres = group['listed_in'].str.split(', ').explode()
  return genres.mode()[0]

genre_by_rating = NF_df.groupby('rating').apply(most_frequent_genre)

# Exibe os resultados
print("Gênero de filme mais comum por classificação indicativa:")
print(genre_by_rating)
```

##Questão 10
***
###Podemos concluir que o filme de gênero Terror melhor avaliado é "Artik". Já o filme de terror pior avaliado é o "The Dentist", de acordo com fontes do Rotten Tomatoes

```python
#Filtramos o gênero dos filmes por "Terror"
terror_df = RT_df.loc[RT_df['genres'] == 'Horror']

nota_alta_terror = terror_df['tomatometer_rating'].max()
nota_baixa_teror = terror_df['tomatometer_rating'].min()

melhor_terror = terror_df.loc[terror_df['tomatometer_rating'] == nota_alta_terror, 'movie_title'].values[0]

pior_terror = terror_df.loc[terror_df['tomatometer_rating'] == nota_baixa_teror, 'movie_title'].values[0]

print(f"O melhor filme do gênero Terror é: {melhor_terror}, com uma nota de {nota_alta_terror}  ")

print("------------------------------------------------")

print(f"O pior filme do gênero Terror é: {pior_terror}, com uma nota de {nota_baixa_teror}  ")
```
